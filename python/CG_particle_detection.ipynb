```matlab
cd('particle_detection');
im = imread('2D_Gaus_sig4.png');whos im;
imshow(im);
```

```matlab
imInv = imcomplement(im);
colormap('gray'), imshow(imInv);
```

```matlab
imBpass = bpass(im,1,15);
figure;imshow(imBpass);
```

```matlab
pk = pkfnd(im,12,15); % brigthness threshold, diameter
subplot(1,2,1)
imshow(im);axis square;
subplot(1,2,2)
scatter(pk(:,1),pk(:,2),10,'filled');box on; axis square; axis([0 size(im,1) 0 size(im,2)])
```

```matlab
cnt = cntrd(imBpass,pk,17);
whos cnt
```

```matlab
scatter(cnt(:,1),cnt(:,2),10,'filled');box on; axis square; axis([0 size(im,1) 0 size(im,2)])

imshow(im);hold on;
scatter(cnt(:,1),cnt(:,2),10,'r*');
```
