```python
from __future__ import division, unicode_literals, print_function  # for compatibility with Python 2 and 3

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
from pandas import DataFrame, Series  # for convenience

import pims
import trackpy as tp

# change the following to %matplotlib notebook for interactive plotting
%matplotlib inline
```

### Import the localization file

```python
data = pd.read_csv("test_data.csv") 
print(data.shape)
```

Show locs data and rename coordinates columns

```python
data.rename(columns={'x_pix': 'x', 'y_pix': 'y'}, inplace=True)
data.head()
```

### Linking positions with radius and gap time

```python
t = tp.link_df(data, 2, memory=3); # Radius in pixel, gap time in frames
```

```python
t.head() # Display result now with additional column 'particle'
```

```python
# Show track length histogram
hist, bins = np.histogram(np.bincount(t.particle.astype(int)),bins=np.arange(30));
plt.step(bins[1:], hist);
plt.gca().set(ylabel='relative frequency', xlabel='track length (frames)');
```

```python
# Filter for track length
t1 = tp.filter_stubs(t, 25) 
# Compare the number of particles in the unfiltered and filtered data.
print('Before:', t['particle'].nunique())
print('After:', t1['particle'].nunique())
```

```python
# Plot the trajectories
plt.figure()
tp.plot_traj(t1);
```

### Mean squared displacement (MSD)

```python
# MSD for each individual trajectory (imsd)
im = tp.imsd(t1, 0.16, 33.3,max_lagtime=9)  # microns per pixel = 0.16, frames per second = 33.3 (30 ms int time)
```

```python
fig, ax = plt.subplots()
ax.plot(im.index, im, 'k-', alpha=0.1)  # black lines, semitransparent
ax.set(ylabel=r'$\langle \Delta r^2 \rangle$ [$\mu$m$^2$]',
       xlabel='lag time $t$')
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
# Display result, one MSD curve per column
im.head()
```

```python
df1 = im.iloc[:,0:1]
df1.head()
```

```python
em = tp.emsd(t1, 0.16, 33.3,max_lagtime=10) # microns per pixel = 100/285., frames per second = 24

fig, ax = plt.subplots()
ax.plot(em.index, em, 'o')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set(ylabel=r'$\langle \Delta r^2 \rangle$ [$\mu$m$^2$]',
       xlabel='lag time $t$')
ax.set(ylim=(8*1e-3, 1e-2));
```

```python
plt.figure()
plt.ylabel(r'$\langle \Delta r^2 \rangle$ [$\mu$m$^2$]')
plt.xlabel('lag time $t$');
tp.utils.fit_powerlaw(em)  # performs linear best fit in log space, plots]
```
